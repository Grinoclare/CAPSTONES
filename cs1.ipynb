```python
# Extracting Channel Info

def extract_channelinfo(channel_id):
    rq=yt.channels().list(
                            part="snippet,ContentDetails,statistics",
                            id=channel_id 
    )
    rs=rq.execute()

    for i in rs['items']:
        dt=dict(Channel_Name=i["snippet"]["title"],
                Channel_ID=i["id"],
                Subscribers_Count=i['statistics']['subscriberCount'],
                Total_Views=i['statistics']['viewCount'],
                Total_Videos=i['statistics']['videoCount'],
                Channel_Description=i["snippet"]["description"],
                Playlist_ID=['contentDetails']['relatedPlaylists']['uploads'])
    return dt
```

```python
# Extracting Video Id 

def extract_videoid(channel_id):
                vd=[]
                rs=yt.channels().list(id=channel_id,
                                part='contentDetails'  
                                        ).execute()
                Playlist_ID=rs['items'][0]['contentDetails']['relatedPlaylists']['uploads'] 

                nextpagetoken=None
                while True :
                        rs1=yt.playlistItems().list(
                                                part='snippet',
                                                playlistId=Playlist_ID,
                                                maxResults=50,
                                                pageToken=nextpagetoken).execute()
                        for i in range(len(rs1['items'])):
                                vd.append(rs1['items'][i]['snippet']['resourceId']['videoId'])
                        nextpagetoken=rs1.get('nextPageToken')
                        if nextpagetoken is None:
                                break
                return vd
```

```python
# Extracting Video Info

def extract_videoinfo(video_id):
                vd1=[]
                for vdid in video_id:
                        rq=yt.videos().list(
                                                part="snippet,ContentDetails,statistics",
                                                id=vdid)
                        rs=rq.execute()

                        for i in rs['items']:
                                dt=dict(Channel_Name=i['snippet']['channelTitle'],
                                        Channel_ID=i['snippet']['channelId'],
                                        Video_ID=i['id'],
                                        Video_Title=i['snippet']['title'],
                                        Tags=i.get('tags'),
                                        Description=i.get('description'),
                                        Published_Date=i['snippet']['publishedAt'],
                                        Duration=i['contentDetails']['duration'],
                                        Views=i.get('viewCount'),
                                        Comments_Count=i.get('commentCount'),
                                        Favourite_Count=i['statistics']['favoriteCount'],
                                        Definition=i['contentDetails']['definition'],
                                        Caption_Status=i['contentDetails']['caption'])   
                                vd1.append(dt)
                return vd1
```
